```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') 

df = pd.read_csv('../data/interim/merged.csv')
data = pd.read_csv('../data/interim/data.csv')
items = pd.read_csv('../data/interim/items.csv')
users = pd.read_csv('../data/interim/users.csv')
```

```python
sns.countplot(x=data['rating'])
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=users['age'])
```

```python
sns.countplot(x=users['gender'])
```

```python
plt.figure(figsize=(5, 8))
sns.countplot(y=users['occupation'])
```

```python
counts = items.sum(axis=0,numeric_only=True)[3:]

sns.barplot(y=counts.index, x=counts.values)
```

```python
sns.histplot(df.groupby('item_id')['rating'].mean(),kde=True)
plt.title("Average film ratings")
```

```python
sns.histplot(df.groupby('user_id')['rating'].mean(),kde=True)
plt.title("User avg. rating scores dist.")
```

```python
sns.histplot(data['user_id'].value_counts(),kde=True)
plt.title("Films rated by user")
```

```python
ratings = {}
genres =  ['unknown', 'Action',
       'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

for g in genres:
    ratings[g] = df.groupby(g)['rating'].mean()[1]
sns.barplot(y=list(ratings.keys()), x=list(ratings.values()))
```
